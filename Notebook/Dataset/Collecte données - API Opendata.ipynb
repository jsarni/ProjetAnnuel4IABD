```python
DATASET = 'horodateurs-transactions-de-paiement'
URL = 'https://opendata.paris.fr/api/records/1.0/search//'
LIMIT = -1

horodateurs_csv = pd.read_csv("horodateurs-mobiliers.csv", sep=';')
horodateurs = [x for x in horodateurs_csv['NÂ° d\'horodateur'].tolist() if x >= 64460203]
horodateurs.sort()

for i in range(len(horodateurs)):
    print('numero {} / {}: {}'.format(i + 1, len(horodateurs), horodateurs[i]))
    filename = 'transactions_horodateur/partition_horodateur/{}.csv'.format(horodateurs[i])
    PARAMS = {
        'dataset': DATASET,
        'refine.horodateur': int(horodateurs[i]),
        'rows': LIMIT
    }

    response = requests.get(url=URL, params=PARAMS)
    data = response.json()

    records =json.dumps(data['records'])
    records_json = pd.read_json(records)
    if(records_json.shape[0] != 0):
        records_fields_list = records_json['fields'].tolist()
        records_fields_json_str = json.dumps(records_fields_list)
        records_df = pd.read_json(records_fields_json_str)
        records_df.to_csv(filename, mode='a', header=False, index=False, sep=';')
```
